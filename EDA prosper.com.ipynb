```python

plt.figure(figsize=(15,6))
plt.title("Paid Loan Amount to the Lender")
plt.hist(prosper_df1.Loan_Amount,alpha=0.4,bins=5);
plt.hist(prosper_df1.Completed_LoanAmount,alpha=0.4,bins=5);
```
